```python
import pandas as pd

# Load train, dev, and test
df_train_full = pd.read_json('/content/drive/MyDrive/Neural Nets project/subtaskA_train_monolingual.jsonl', lines=True)
df_dev = pd.read_json('/content/drive/MyDrive/Neural Nets project/subtaskA_dev_final_monolingual.jsonl', lines=True)
df_test = pd.read_json('/content/drive/MyDrive/Neural Nets project/subtaskA_test_monolingual.jsonl', lines=True)

# Label distribution check
print("Train:", df_train_full['label'].value_counts().to_dict())
print("Dev:", df_dev['label'].value_counts().to_dict())
print("Test:", df_test['label'].value_counts().to_dict())

# Create combined dataset for fine-tuning pre-classifier
df_combined = pd.concat([df_train_full, df_dev], ignore_index=True)

# Sample only 10% of data for pre-classifier fine-tuning
df_train_sample = df_combined.groupby('label', group_keys=False).apply(lambda x: x.sample(frac=0.10, random_state=42)).reset_index(drop=True)
print("Sampled Fine-tune Data:", df_train_sample['label'].value_counts())
```

```python
from transformers import AutoTokenizer
from datasets import Dataset

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("openai-community/roberta-base-openai-detector")

# Preprocessing function
def preprocess_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=512)

# Prepare tokenized datasets
train_dataset = Dataset.from_pandas(df_train_sample[['text', 'label']])
dev_dataset = Dataset.from_pandas(df_dev[['text', 'label']])

tokenized_train = train_dataset.map(preprocess_function, batched=True)
tokenized_dev = dev_dataset.map(preprocess_function, batched=True)

tokenized_train = tokenized_train.rename_column("label", "labels")
tokenized_dev = tokenized_dev.rename_column("label", "labels")

tokenized_train.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
tokenized_dev.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
```

```python
import os
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
import evaluate
import numpy as np

# Load model
model = AutoModelForSequenceClassification.from_pretrained(
    "openai-community/roberta-base-openai-detector",
    num_labels=2
)

# Evaluation metrics
accuracy = evaluate.load("accuracy")
f1 = evaluate.load("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    acc = accuracy.compute(predictions=predictions, references=labels)
    f1_score = f1.compute(predictions=predictions, references=labels, average='macro')
    return {
        "accuracy": acc["accuracy"],
        "macro_f1": f1_score["f1"]
    }

# TrainingArguments
training_args = TrainingArguments(
    output_dir="./roberta-openai-detector-finetuned",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=32,
    num_train_epochs=3,
    weight_decay=0.01,
    load_best_model_at_end=True,
    metric_for_best_model="macro_f1",
    greater_is_better=True,
    logging_dir="./logs",
    fp16=True,
    report_to="none"
)

# Trainer setup
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_dev,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

# ✅ Start training
trainer.train()
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
import pandas as pd
import numpy as np

# ✅ Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ✅ Load tokenizer and model
model_path = "/content/roberta-openai-detector-finetuned"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path, output_hidden_states=True)
model.eval().to(device)

# ✅ Reload all three datasets
df_train = pd.read_json('/content/drive/MyDrive/Neural Nets project/subtaskA_train_monolingual.jsonl', lines=True)
df_dev = pd.read_json('/content/drive/MyDrive/Neural Nets project/subtaskA_dev_final_monolingual.jsonl', lines=True)
df_test = pd.read_json('/content/drive/MyDrive/Neural Nets project/subtaskA_test_monolingual.jsonl', lines=True)

# ✅ Dataset Class
class TextDataset(Dataset):
    def __init__(self, texts, labels=None):
        self.texts = texts
        self.labels = labels

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = tokenizer(
            self.texts[idx],
            truncation=True,
            padding='max_length',
            max_length=512,
            return_tensors="pt"
        )
        item = {key: val.squeeze(0) for key, val in encoding.items()}
        if self.labels is not None:
            item['labels'] = torch.tensor(self.labels[idx])
        return item

# ✅ Feature Extraction Function
def extract_features(df, name):
    dataset = TextDataset(df["text"].tolist(), df["label"].tolist())
    loader = DataLoader(dataset, batch_size=32)

    logits_list = []
    cls_hidden_states = []

    with torch.no_grad():
        for batch in tqdm(loader, desc=f"Extracting features from {name}"):
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            hidden_states = outputs.hidden_states[-1][:, 0, :]  # CLS token embedding

            logits_list.append(logits.cpu().numpy())
            cls_hidden_states.append(hidden_states.cpu().numpy())

    # Concatenate and save
    logits_all = np.concatenate(logits_list, axis=0)
    cls_all = np.concatenate(cls_hidden_states, axis=0)

    # Save logits as CSV with labels
    logits_df = pd.DataFrame(logits_all, columns=["logit_0", "logit_1"])
    logits_df["label"] = df["label"].values
    logits_df.to_csv(f"/content/{name}_logits.csv", index=False)

    # Save CLS embeddings as .npy
    np.save(f"/content/{name}_cls_hidden.npy", cls_all)

    print(f"✅ Done: {name} → logits: {logits_all.shape}, CLS: {cls_all.shape}")
    return logits_df, cls_all

# ✅ Run for train, dev, and test
logits_train, cls_train = extract_features(df_train, "train")
logits_dev, cls_dev = extract_features(df_dev, "dev")
logits_test, cls_test = extract_features(df_test, "test")
```

```python
7# 🛠 INSTALLS
!pip install -q textstat umap-learn tqdm
!pip install -q nltk transformers

# 📦 IMPORTS
import os
import time
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import umap
from textstat import flesch_reading_ease
from nltk import word_tokenize, pos_tag
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from transformers import pipeline
import nltk

# ✅ NLTK DATA
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
nltk.download("averaged_perceptron_tagger_eng")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("punkt_tab")  # 🚨 Fix for past crashes

# ✅ Load logits and hidden states (from Step 4 outputs)
logits_train = np.load("logits_train.npy")
logits_dev   = np.load("logits_dev.npy")
logits_test  = np.load("logits_test.npy")

cls_train = np.load("cls_train.npy")
cls_dev   = np.load("cls_dev.npy")
cls_test  = np.load("cls_test.npy")


# ⏳ Dimensionality Reduction
def reduce_dims(cls_vectors):
    scaler = StandardScaler()
    cls_scaled = scaler.fit_transform(cls_vectors)

    pca = PCA(n_components=2, random_state=42)
    pca_reduced = pca.fit_transform(cls_scaled)

    reducer = umap.UMAP(n_components=2, n_neighbors=100, min_dist=0.01, metric="jaccard", random_state=42)
    umap_reduced = reducer.fit_transform(cls_scaled)

    return pca_reduced, umap_reduced

cls_all = np.vstack([cls_train, cls_dev, cls_test])
pca_all, umap_all = reduce_dims(cls_all)
pca_train, pca_dev, pca_test = np.split(pca_all, [len(cls_train), len(cls_train)+len(cls_dev)])
umap_train, umap_dev, umap_test = np.split(umap_all, [len(cls_train), len(cls_train)+len(cls_dev)])

# ✅ GPU Sentiment + Emotion Pipelines
sentiment_pipe = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english", device=0)
emotion_pipe = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True, device=0)

# 📌 Feature extraction setup
stop_words = set(stopwords.words("english"))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# 🔍 Extract features per row
def extract_text_features(text):
    try:
        tokens = word_tokenize(text)
        words = [w for w in tokens if w.isalpha()]
        num_words = len(words) if len(words) > 0 else 1

        pos_tags = pos_tag(words)
        num_pronouns = sum(1 for _, t in pos_tags if t in ["PRP", "PRP$", "WP"])
        num_conj = sum(1 for _, t in pos_tags if t.startswith("CC"))
        num_adv = sum(1 for _, t in pos_tags if t.startswith("RB"))

        stems = [stemmer.stem(w) for w in words]
        lemmas = [lemmatizer.lemmatize(w) for w in words]

        readability = flesch_reading_ease(text)

        sentiment = sentiment_pipe(text[:512])[0]
        emotion_scores = emotion_pipe(text[:512])[0]

        emotion_dict = {e['label'].lower(): e['score'] for e in emotion_scores}
        emotion_dict = {k: emotion_dict.get(k, 0.0) for k in ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'surprise']}

        return {
            "num_tokens": len(tokens),
            "num_words": num_words,
            "ratio_pronouns": num_pronouns / num_words,
            "ratio_conjunctions": num_conj / num_words,
            "ratio_adverbs": num_adv / num_words,
            "readability": readability,
            "sentiment_score": sentiment["score"] if sentiment["label"] == "POSITIVE" else -sentiment["score"],
            **emotion_dict
        }
    except Exception as e:
        print(f"⚠️ Skipping due to error: {e}")
        return None

# 🧱 Batch feature builder with ETA tracking
def build_feature_dataframe(df, logits, pca, umap, name="dataset"):
    print(f"\n🚀 Building features for {name} — {len(df)} rows")
    start = time.time()
    rows = []
    for i, row in tqdm(df.iterrows(), total=len(df)):
        feats = extract_text_features(row["text"])
        if feats is None:
            continue
        feats["logit_0"] = logits[i][0]
        feats["logit_1"] = logits[i][1]
        feats["pca_0"] = pca[i][0]
        feats["pca_1"] = pca[i][1]
        feats["umap_0"] = umap[i][0]
        feats["umap_1"] = umap[i][1]
        feats["label"] = row.get("label", -1)
        rows.append(feats)
    df_out = pd.DataFrame(rows)
    print(f"✅ Finished {name} — Time: {(time.time() - start) / 60:.2f} minutes | Final rows: {len(df_out)}")
    return df_out

# 🔁 Build features
df_feat_train = build_feature_dataframe(df_train_full, logits_train, pca_train, umap_train, "Train")
df_feat_dev   = build_feature_dataframe(df_dev, logits_dev, pca_dev, umap_dev, "Dev")
df_feat_test  = build_feature_dataframe(df_test, logits_test, pca_test, umap_test, "Test")

# 💾 Save features
df_feat_train.to_csv("features_train.csv", index=False)
df_feat_dev.to_csv("features_dev.csv", index=False)
df_feat_test.to_csv("features_test.csv", index=False)
print("✅ Step 5 COMPLETE: Feature CSVs saved.")
```

```python
# 📦 Install dependencies (run this only if not already installed)
!pip install scikit-learn pandas

# 📚 Imports
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
import time

# ⏱ ETA Tracker
def eta(start, percent):
    elapsed = time.time() - start
    total_est = elapsed / percent
    remaining = total_est - elapsed
    print(f"⏳ {percent*100:.2f}% done | Elapsed: {elapsed:.2f}s | ETA: {remaining:.2f}s")

start_time = time.time()

# 📁 Load extracted features from Step 5
df_train = pd.read_csv("features_train.csv")
df_dev = pd.read_csv("features_dev.csv")
df_test = pd.read_csv("features_test.csv")

# 🎯 Split training data by pre-classifier prediction
threshold = 0.5  # This was binary, so logit_1 > 0.5 ⇒ "machine", else "human"
df_train_human = df_train[df_train["logit_1"] <= threshold]
df_train_machine = df_train[df_train["logit_1"] > threshold]

# 🎯 Similarly split dev set to evaluate both MLPs separately
df_dev_human = df_dev[df_dev["logit_1"] <= threshold]
df_dev_machine = df_dev[df_dev["logit_1"] > threshold]

# 🎯 Features to train on
features_to_use = [col for col in df_train.columns if col not in ["label"]]

# 🧪 Train MLP for "machine" predictions
scaler_m = StandardScaler()
X_m_train = scaler_m.fit_transform(df_train_machine[features_to_use])
y_m_train = df_train_machine["label"]

mlp_machine = MLPClassifier(hidden_layer_sizes=(64,), max_iter=500, random_state=42)
mlp_machine.fit(X_m_train, y_m_train)

X_m_dev = scaler_m.transform(df_dev_machine[features_to_use])
y_m_dev = df_dev_machine["label"]
y_m_pred = mlp_machine.predict(X_m_dev)

print("🧠 Correction MLP for 'machine' predictions:")
print(classification_report(y_m_dev, y_m_pred))
print("Accuracy:", accuracy_score(y_m_dev, y_m_pred))
eta(start_time, 0.5)

# 🧪 Train MLP for "human" predictions
scaler_h = StandardScaler()
X_h_train = scaler_h.fit_transform(df_train_human[features_to_use])
y_h_train = df_train_human["label"]

mlp_human = MLPClassifier(hidden_layer_sizes=(64,), max_iter=500, random_state=42)
mlp_human.fit(X_h_train, y_h_train)

X_h_dev = scaler_h.transform(df_dev_human[features_to_use])
y_h_dev = df_dev_human["label"]
y_h_pred = mlp_human.predict(X_h_dev)

print("🧠 Correction MLP for 'human' predictions:")
print(classification_report(y_h_dev, y_h_pred))
print("Accuracy:", accuracy_score(y_h_dev, y_h_pred))
eta(start_time, 1.0)

# 💾 Save trained MLPs and scalers
import joblib
joblib.dump(mlp_machine, "mlp_machine.pkl")
joblib.dump(mlp_human, "mlp_human.pkl")
joblib.dump(scaler_m, "scaler_machine.pkl")
joblib.dump(scaler_h, "scaler_human.pkl")

print("✅ Step 6 complete: Correction MLPs trained and saved.")
```

```python
# 📦 Import all necessary modules
import pandas as pd
import numpy as np
import joblib
import time
from sklearn.metrics import classification_report, accuracy_score

# ⏱ ETA Tracker
def eta(start, percent):
    elapsed = time.time() - start
    total_est = elapsed / percent if percent > 0 else 0
    remaining = total_est - elapsed
    print(f"⏳ {percent*100:.2f}% done | Elapsed: {elapsed:.2f}s | ETA: {remaining:.2f}s")

start_time = time.time()

# ✅ Load test data and models
df_test = pd.read_csv("features_test.csv")
mlp_machine = joblib.load("mlp_machine.pkl")
mlp_human = joblib.load("mlp_human.pkl")
scaler_m = joblib.load("scaler_machine.pkl")
scaler_h = joblib.load("scaler_human.pkl")

# 🎯 Identify feature columns
features_to_use = [col for col in df_test.columns if col not in ["label"]]

# ✅ Use pre-classifier's logits to route through the right MLP
pre_preds = df_test[["logit_0", "logit_1"]].values.argmax(axis=1)
true_labels = df_test["label"].values

# 🔁 Correction loop
final_preds = []
for i, row in df_test.iterrows():
    feats_df = pd.DataFrame([row[features_to_use].values], columns=features_to_use)
    if pre_preds[i] == 1:
        corrected = mlp_machine.predict(scaler_m.transform(feats_df))[0]
    else:
        corrected = mlp_human.predict(scaler_h.transform(feats_df))[0]
    final_preds.append(corrected)

# 🧾 Final Metrics
eta(start_time, 1.0)
report = classification_report(true_labels, final_preds, digits=4)
accuracy = accuracy_score(true_labels, final_preds)

print("\n✅ Step 7 Complete — Final Evaluation")
print(report)
print("Accuracy:", round(accuracy * 100, 2), "%")
```

```python
# 📦 Install missing dependencies (if needed)
!pip install scikit-learn pandas matplotlib textstat transformers umap-learn

# 📚 Imports
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.decomposition import PCA
import umap
import joblib
import nltk
from nltk import word_tokenize, pos_tag
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from textstat import flesch_reading_ease
from transformers import pipeline
import matplotlib.pyplot as plt
import time

# ⏱ ETA tracking
def eta(start, pct):
    elapsed = time.time() - start
    remaining = (elapsed / pct) - elapsed
    print(f"⏳ {pct*100:.2f}% done | Elapsed: {elapsed:.2f}s | ETA: {remaining:.2f}s")

start = time.time()

# ✅ Download required NLTK packages
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')
nltk.download('punkt_tab')

# ✅ Load the test data
df_newtest = pd.read_csv("/Monolingual_crossdomain_testdata.csv")

# ✅ Load trained MLPs and scalers
mlp_machine = joblib.load("mlp_machine.pkl")
mlp_human   = joblib.load("mlp_human.pkl")
scaler_m    = joblib.load("scaler_machine.pkl")
scaler_h    = joblib.load("scaler_human.pkl")

# ✅ Load tokenizer & classifier for sentiment/emotion
sentiment_pipe = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english", device=0)
emotion_pipe   = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True, device=0)

# ⚙️ NLP Tools
stop_words = set(stopwords.words("english"))
stemmer    = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# 🧠 Feature extraction from text
def extract_text_features(text):
    try:
        tokens = word_tokenize(text)
        words = [w for w in tokens if w.isalpha()]
        num_words = len(words) if len(words) > 0 else 1

        pos_tags = pos_tag(words)
        num_pronouns = sum(1 for _, t in pos_tags if t in ["PRP", "PRP$", "WP"])
        num_conj = sum(1 for _, t in pos_tags if t.startswith("CC"))
        num_adv = sum(1 for _, t in pos_tags if t.startswith("RB"))

        stems = [stemmer.stem(w) for w in words]
        lemmas = [lemmatizer.lemmatize(w) for w in words]

        readability = flesch_reading_ease(text)
        sentiment = sentiment_pipe(text[:512])[0]
        emotion_scores = emotion_pipe(text[:512])[0]

        emotion_dict = {e['label'].lower(): e['score'] for e in emotion_scores}
        emotion_dict = {k: emotion_dict.get(k, 0.0) for k in ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'surprise']}

        return {
            "num_tokens": len(tokens),
            "num_words": num_words,
            "ratio_pronouns": num_pronouns / num_words,
            "ratio_conjunctions": num_conj / num_words,
            "ratio_adverbs": num_adv / num_words,
            "readability": readability,
            "sentiment_score": sentiment["score"] if sentiment["label"] == "POSITIVE" else -sentiment["score"],
            **emotion_dict
        }
    except Exception as e:
        print("⚠️ Skipping due to error:", e)
        return None

# ✅ Extract features
features = []
for i, row in df_newtest.iterrows():
    feats = extract_text_features(row["text"])
    if feats is None: continue
    feats["logit_0"] = -1  # placeholder
    feats["logit_1"] = 1 if row["label"] == 1 else 0
    feats["pca_0"] = 0
    feats["pca_1"] = 0
    feats["umap_0"] = 0
    feats["umap_1"] = 0
    feats["label"] = row["label"]
    features.append(feats)
    if (i+1) % 2 == 0: eta(start, (i+1)/len(df_newtest))

df_feat_new = pd.DataFrame(features)

# ✅ Predict with appropriate correction MLP
X_machine = df_feat_new[df_feat_new["logit_1"] > 0.5]
X_human   = df_feat_new[df_feat_new["logit_1"] <= 0.5]

X_m_scaled = scaler_m.transform(X_machine.drop(columns=["label"]))
X_h_scaled = scaler_h.transform(X_human.drop(columns=["label"]))

y_m_pred = mlp_machine.predict(X_m_scaled)
y_h_pred = mlp_human.predict(X_h_scaled)

# ✅ Reconstruct full prediction array
y_true = df_feat_new["label"].values
y_pred = np.zeros(len(df_feat_new), dtype=int)

y_pred[X_machine.index] = y_m_pred
y_pred[X_human.index]   = y_h_pred

# ✅ Metrics and confusion matrix
print("\n🎯 Cross-Domain Evaluation Results:")
print(classification_report(y_true, y_pred, digits=4))
print("✅ Accuracy:", accuracy_score(y_true, y_pred))

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Human", "Machine"])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix - Cross-Domain Test")
plt.show()
```

```python
import pandas as pd
import json

def load_jsonl_safe(filepath):
    data = []
    with open(filepath, "r", encoding="utf-8") as f:
        for i, line in enumerate(f):
            try:
                data.append(json.loads(line.strip()))
            except json.JSONDecodeError as e:
                print(f"⚠️ Skipped malformed line {i}: {e}")
    return pd.DataFrame(data)

# Load the full train and dev datasets
df_train = load_jsonl_safe("subtaskA_train_multilingual.jsonl")
df_dev_full = load_jsonl_safe("subtaskA_dev_multilingual.jsonl")

# Split dev: 2500 for validation, 1500 for in-domain test
df_val = df_dev_full.iloc[:2500].reset_index(drop=True)
df_test = df_dev_full.iloc[2500:].reset_index(drop=True)

# Confirm sizes
print(f"✅ Train size: {len(df_train)}")
print(f"✅ Validation size: {len(df_val)}")
print(f"✅ Test size: {len(df_test)}")
```

```python
import re

def clean_text(text):
    text = re.sub(r"\s+", " ", text)  # Replace multiple whitespace/newlines with single space
    text = text.strip()
    return text

# Apply cleaning
df_train["text"] = df_train["text"].astype(str).apply(clean_text)
df_val["text"] = df_val["text"].astype(str).apply(clean_text)
df_test["text"] = df_test["text"].astype(str).apply(clean_text)

# Drop rows with missing text or label (precautionary)
df_train.dropna(subset=["text", "label"], inplace=True)
df_val.dropna(subset=["text", "label"], inplace=True)
df_test.dropna(subset=["text", "label"], inplace=True)

# Save to CSVs (no index)
df_train[["text", "label"]].to_csv("train.csv", index=False)
df_val[["text", "label"]].to_csv("val.csv", index=False)
df_test[["text", "label"]].to_csv("test.csv", index=False)

print("✅ Saved train.csv, val.csv, and test.csv")
```

```python
from transformers import AutoTokenizer
from datasets import Dataset
import pandas as pd

# Load CSVs
df_train = pd.read_csv("train.csv")
df_val = pd.read_csv("val.csv")
df_test = pd.read_csv("test.csv")

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-base")

# Tokenization function
def tokenize_function(batch):
    texts = [str(x) if x is not None else "" for x in batch["text"]]
    return tokenizer(
        texts,
        padding="max_length",
        truncation=True,
        max_length=512,
    )



# Convert DataFrames to Hugging Face Datasets
train_ds = Dataset.from_pandas(df_train)
val_ds = Dataset.from_pandas(df_val)
test_ds = Dataset.from_pandas(df_test)

# Tokenize datasets
train_ds = train_ds.map(tokenize_function, batched=True)
val_ds = val_ds.map(tokenize_function, batched=True)
test_ds = test_ds.map(tokenize_function, batched=True)

# Remove unused columns (if they exist)
for col in ["text", "__index_level_0__"]:
    if col in train_ds.column_names:
        train_ds = train_ds.remove_columns(col)
    if col in val_ds.column_names:
        val_ds = val_ds.remove_columns(col)
    if col in test_ds.column_names:
        test_ds = test_ds.remove_columns(col)

# Format datasets for PyTorch
train_ds.set_format("torch")
val_ds.set_format("torch")
test_ds.set_format("torch")

print("✅ Tokenization and formatting completed.")


print("✅ Tokenized datasets are ready: train_ds, val_ds, test_ds")
```

```python
from transformers import TrainingArguments, Trainer, EarlyStoppingCallback, AutoTokenizer
from transformers import AutoModelForSequenceClassification
import evaluate
import numpy as np
from transformers import DataCollatorWithPadding

# Load model
tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-base")
model = AutoModelForSequenceClassification.from_pretrained("xlm-roberta-base", num_labels=2)

# Evaluation metrics
accuracy = evaluate.load("accuracy")
f1 = evaluate.load("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy.compute(predictions=predictions, references=labels)["accuracy"],
        "f1": f1.compute(predictions=predictions, references=labels, average="macro")["f1"]
    }

# Data collator
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

# Training args - Set A
training_args_A = TrainingArguments(
    output_dir="./xlm_base_run_A",
    eval_strategy="epoch",
    save_strategy="epoch",
    save_total_limit=2,
    learning_rate=5e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    num_train_epochs=3,
    weight_decay=0.01,
    warmup_ratio=0.1,
    load_best_model_at_end=True,
    metric_for_best_model="eval_f1",
    greater_is_better=True,
    logging_dir="./logs_A",
    logging_strategy="epoch",
    fp16=True,
    report_to="none"
)

# Trainer A with early stopping
trainer_A = Trainer(
    model=model,
    args=training_args_A,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=1)]
)

# Train
trainer_A.train()
```

```python
# Save best model, tokenizer, and training state (Set A)
trainer_A.save_model()  # Saves to training_args_A.output_dir
tokenizer.save_pretrained(trainer_A.args.output_dir)

# Optionally save training state
trainer_A.state.save_to_json(f"{trainer_A.args.output_dir}/trainer_state.json")

print(f"✅ Model, tokenizer, and state saved to: {trainer_A.args.output_dir}")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model_path = "./xlm_base_run_A"

tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)
```

```python
import pandas as pd
from datasets import Dataset

# Load the full dev set
df_dev = pd.read_json("subtaskA_dev_multilingual.jsonl", lines=True)

# Select the last 1,500 rows for in-domain test set
df_test = df_dev.iloc[-1500:].reset_index(drop=True)
test_ds = Dataset.from_pandas(df_test)

# Tokenize
def tokenize_function(example):
    return tokenizer(example["text"], truncation=True)

test_ds = test_ds.map(tokenize_function, batched=True)
test_ds = test_ds.remove_columns([col for col in test_ds.column_names if col not in ["input_ids", "attention_mask", "label"]])
test_ds.set_format("torch")
```

```python
import evaluate
import numpy as np
from transformers import Trainer, DataCollatorWithPadding
import os
os.environ["WANDB_DISABLED"] = "true"


accuracy = evaluate.load("accuracy")
f1 = evaluate.load("f1")

def compute_metrics(pred):
    logits, labels = pred
    preds = np.argmax(logits, axis=1)
    return {
        "accuracy": accuracy.compute(predictions=preds, references=labels)["accuracy"],
        "f1": f1.compute(predictions=preds, references=labels, average="macro")["f1"]
    }

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

trainer = Trainer(
    model=model,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# Evaluate
metrics = trainer.evaluate(eval_dataset=test_ds)
print("📊 In-domain test set evaluation:")
print(metrics)
```
